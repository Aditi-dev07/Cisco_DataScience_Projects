# Volcano Eruptions

Which volcanoes have experienced the longest eruptions?

The dataset `volcanic-eruptions.csv` includes the start and end dates for all volcanic eruptions since 1800, along with each volcano's identification number. For eruptions that are still ongoing as of December 2024, the end date is recorded as December 2024.

This dataset excludes any eruptive pauses shorter than three months. If an eruption resumes after more than three months of inactivity, it is classified as a new eruption.

Note: Volcano Yasur in Vanuatu is not included in this list due to the absence of a clear start date.

```python
import pandas as pd
import matplotlib.pyplot as plt

eruptions = pd.read_csv('volcanic-eruptions.csv')
eruptions.head(3)
```

### Additional dataset

The dataset `volcano-list.csv` provides detailed information about each volcano, including its name, country, latitude, longitude, and type.

```python
volcanoes = pd.read_csv('volcano-list.csv')
volcanoes.head(3)
```

### Project Ideas:

- Find the volcanoes that were erupting as of Dec 2024.

- Find the volcanoes that have had the longest volcanic eruptions.

Hints:
- Use `pd.to_datetime`.

- Merge the volcanoes dataframe into eruptions.

- Before the merge, reduce the dataframes to the columns of interest.

- Use `df.sort_values`.

```python
eruptions['start_date'] = pd.to_datetime(eruptions['start_date'])
eruptions['end_date'] = pd.to_datetime(eruptions['end_date'])
eruptions.head()
```

```python
mergedf = eruptions.merge(volcanoes, on= 'volcano_id')
mergedf.head()
```

```python
mergedf.columns
```

```python
mergedf.drop(columns=['volcano_id','volcanic_region_group','volcanic_region',
                           'volcano_landform','primary_volcano_type','activity_evidence',
                           'dominant_rock_type','latitude', 'longitude'],inplace=True)
mergedf.head()
```

```python
mergedf.sort_values(by="start_date", ascending= False)
mergedf.head()
```

##1.Find the volcanoes that were erupting as of Dec 2024.

```python
cutoff_date = pd.Timestamp("2024-12-01")

erupting_dec_2024 = mergedf[
    (mergedf["start_date"] <= cutoff_date) &
    (
        mergedf["end_date"].isna() |
        (mergedf["end_date"] >= cutoff_date)
    )
]

print("Volcanoes erupting as of December 2024:")
display(erupting_dec_2024[["volcano_name", "start_date", "end_date"]].drop_duplicates().sort_values(by="volcano_name"))
```

##2.Find the volcanoes that have had the longest volcanic eruptions.

```python
mergedf['duration_days'] = (mergedf['end_date'] - mergedf['start_date']).dt.days

longest_eruptions = mergedf.sort_values(by='duration_days', ascending=False)

print("Volcanoes with the longest eruptions:")
display(longest_eruptions[["volcano_name", "start_date", "end_date", "duration_days"]].head(10))
```
