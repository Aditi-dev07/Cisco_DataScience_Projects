### Making sense of the data

As you are cleaning a dataframe, it's important to understand what the values in each row and column mean.

Here each row represents one coffee enthusiast and their yes/no answers to each survey question. If the enthusiast adds a particular type of dairy to their coffee, that column will show a value of 1, while a value of 0 indicates that they do not use that type of dairy.

Notice that there are also several `NaN` values (Not a Number), which represents missing values. In this case, NaNs indicate that the survey question was not answered.

Let's use the `isna()` method to take a closer look at how many `NaN` values each column has.

Activity Goals:
- Create a boolean dataframe with True for NaNs and False otherwise.
- Add up the True values per column to get the number of NaNs.


```python
dairy.isna().sum()
```

Notice that each dairy option has the same number of NaN values. Why might this be the case?

This particular survey had a question that asks if you add any dairy to your coffee. If the respondent answered "No", the dairy part of the survey was skipped entirely.

Since these rows have no data, we can drop them using `dropna()`.

Activity Goals:
- Remove rows with NaN values.
- Display the remaining rows.


```python
dairy = dairy.dropna()
dairy #display dataframe
```

After dropping all the rows with NaNs, there are now 523 rows left in the dataframe. Many rows were removed, such as the row with index 1.

Now that we have cleaned the data, we can investigate dairy preferences.

### What dairy options are most preferred?

What percentage of people consume each type of dairy? We can use the `mean()` method on the dataframe to calculate the ratio of "yes" answers for each column. Multiplying by 100 will give us these ratios as percentages.

Activity Goals:
- Calculate the percentage for each dairy.
- Display the result.

```python
dairy_preferences = dairy.mean() * 100
dairy_preferences #display results
```

It looks like whole milk is the most popular. A quick scan of these numbers tells us that the percentages add up to more than 100%. This happens because people often like more than one type of dairy in their coffee.

### Visualizing dairy preferences
Let's visualize the dairy preferences as a horizontal bar chart. We need to first sort `dairy_preferences` and then will use the `barh()` method from matplotlib.

Activity Goals:
- Sort diary_preferences.
- Display the series.

```python
dairy_preferences = dairy_preferences.sort_values()
dairy_preferences #display series
```

And now we can create our horizontal bar plot.

Activity Goals:
- Import matplotlib.
- Make a horizontal bar chart of dairy preferences.
- Label the x-axis.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 3))
plt.barh(dairy_preferences.index, dairy_preferences)
plt.xlabel("Percent")
```

The most popular plant based dairy choice is oat milk. This might be a great choice for the default dairy option in a specialty coffee shop.

# Task
Analyze the "CoffeeSurvey" dataset to determine which age groups consume the most coffee. This involves selecting relevant columns, converting categorical coffee consumption data to a numerical format, calculating the average coffee consumption for each age group, and visualizing these averages in a horizontal bar chart to identify trends.

## Prepare Data for Analysis

### Subtask:
Select the 'What is your age?' and 'How many cups of coffee do you typically drink per day?' columns from the `CoffeeSurvey` DataFrame and remove any rows where these columns have missing values. This ensures a clean dataset for our analysis.


```python
coffee_consumption_data = CoffeeSurvey[['What is your age?', 'How many cups of coffee do you typically drink per day?']]
coffee_consumption_data = coffee_consumption_data.dropna()
coffee_consumption_data.head()
```


To analyze coffee consumption by age group, I need to convert the 'How many cups of coffee do you typically drink per day?' column from categorical strings to numerical values. This mapping will assign a central numerical value to each category, allowing for quantitative analysis.



```python
coffee_consumption_data['Coffee Cups per Day'] = coffee_consumption_data['How many cups of coffee do you typically drink per day?'].map({
    'Less than 1': 0.5,
    '1': 1,
    '2': 2,
    '3': 3,
    '4': 4,
    '5 or more': 5
})
coffee_consumption_data.head()
```


Now that the coffee consumption data has been converted to numerical values, I will calculate the average coffee consumption for each age group using the `groupby()` method and `mean()` aggregation. This will summarize the coffee consumption by age.



```python
average_coffee_per_age = coffee_consumption_data.groupby('What is your age?')['Coffee Cups per Day'].mean().reset_index()
average_coffee_per_age = average_coffee_per_age.rename(columns={'What is your age?': 'Age Group', 'Coffee Cups per Day': 'Average Coffee Cups per Day'})
print(average_coffee_per_age)
```


To better visualize the trends in coffee consumption across age groups, I will sort the `average_coffee_per_age` DataFrame by 'Average Coffee Cups per Day' in ascending order. This will arrange the age groups from least to most coffee consumed, making the visualization more interpretable.



```python
average_coffee_per_age = average_coffee_per_age.sort_values(by='Average Coffee Cups per Day', ascending=True)
print(average_coffee_per_age)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 7))
plt.bar(average_coffee_per_age['Age Group'], average_coffee_per_age['Average Coffee Cups per Day'], color='lightcoral')
plt.xlabel('Age Group')
plt.ylabel('Average Coffee Cups per Day')
plt.title('Average Coffee Consumption by Age Group')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add numerical labels to each bar
for index, value in enumerate(average_coffee_per_age['Average Coffee Cups per Day']):
    plt.text(index, value, f'{value:.2f}', ha='center', va='bottom') # Adjust position for vertical bars

plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
