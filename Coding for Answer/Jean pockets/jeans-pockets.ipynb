# Jean pockets

Are women jean pockets ridiculously small? Let's look at the data!

The file `jean-pocket-measurements.csv` shows pocket measurements for 20 popular brands.
Four pairs of jeans from each brand were measured: men's and women's skinny and straight styles. All jeans were designated a 32-inch waistband.

The following measurements are included:
- `height_front`: maximum front pocket height (cm)
- `height_back`: maximum back pocket height (cm)
- `width_front`: maximum front pocket width (cm)
- `width_back`: maximum back pocket width (cm)

```python
import pandas as pd
df = pd.read_csv('jean-pocket-measurements.csv')
df.head()
```

### Project Ideas

- What is the average difference in pocket `height_front` between women's and men's jeans?

- Is there a significant difference in pocket `height_front` between skinny and straight styles within the same gender?

- How do back pocket sizes compare between women's and men's jeans?

- Measure your phone's height. What percentage of women's and men's jeans can comfortably fit your phone in the pockets?



What is the average difference in pocket height_front between women's and men's jeans?

```python
men_avg = df.query('gender == "men"')['height_front'].mean()
women_avg = df.query('gender == "women"')['height_front'].mean()
print(men_avg)
print(women_avg)
```

```python
difference_avg = men_avg - women_avg.round(2)
print(difference_avg)
```

Is there a significant difference in pocket height_front between skinny and straight styles within the same gender?

```python
men_skinny_avg = df.query('gender == "men" & style == "skinny"')['height_front'].mean()
men_straight_avg = df.query('gender == "men" & style == "straight"')['height_front'].mean()
print(f"Men's Skinny average front pocket height: {men_skinny_avg:.2f}")
print(f"Men's Straight average front pocket height: {men_straight_avg:.2f}")

avg_diff_men = men_skinny_avg - men_straight_avg.round()
print(f"Difference: {avg_diff:.2f}")

women_skinny_avg = df.query('gender == "women" & style == "skinny"')['height_front'].mean()
women_straight_avg = df.query('gender == "women" & style == "straight"')['height_front'].mean()
print(f"Women's Skinny average front pocket height: {women_skinny_avg:.2f}")
print(f"Women's Straight average front pocket height: {women_straight_avg:.2f}")

avg_diff_women = women_skinny_avg - women_straight_avg.round()
print(f"Difference: {avg_diff_women:.2f}")
```


How do back pocket sizes compare between women's and men's jeans?

```python
dw = df.query('gender == "women"')['width_back'].mean()
dwh= df.query('gender == "women"')['height_back'].mean()
dw_area = dwh * dw

print(f"Women back pocket size :{dw_area} ")

dm = df.query('gender == "men"')['width_back'].mean()
dmh = df.query('gender == "men"')['height_back'].mean()
dm_area = dmh * dm

print(f"Men back pocket size: {dm_area}")
```

Measure your phone's height. What percentage of women's and men's jeans can comfortably fit your phone in the pockets

My phones's height = 16.57 cm

```python
phone_height = 16.57
women_hp= df.query('gender == "women"')['height_front'].mean()
women_percentage = (women_hp / phone_height) * 100

men_hp = df.query('gender == "men"')['height_front'].mean()
men_percentage = (men_hp / phone_height) * 100

print(f"Percentage of Women Pockets that fit the phone: {women_percentage:.2f}%")
print(f"Percentage of Men Pockets that fit the phone: {men_percentage:.2f}%")
```
