```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('largest-islands.csv')
df.head()
```

### Project Ideas

- What are the 10 largest islands in the tropics?

- What are the largest islands in each `region`?

- Create a line graph with `area` on the y-axis and `rank` on the x-axis. The data should be ordered by `rank`, from largest to smallest.

- What islands are composed of multiple countries?
	- Hint: try `countries.str.contains(",")`

```python
df.query("climate == 'tropics'").nlargest(10, 'area')
```

```python
largest_islands_by_region = df.loc[df.groupby('region')['area'].idxmax()]
display(largest_islands_by_region.sort_values(by='area', ascending=False))
```

```python
import matplotlib.pyplot as plt

df_sorted_by_rank = df.sort_values(by='rank', ascending=True)
df_sorted_by_rank.plot(x='rank', y='area', kind='line')
plt.title('Island Area by Rank')
plt.xlabel('Rank')
plt.ylabel('Area (km)')
plt.grid(True)
plt.show()
```

```python
df_multiple_countries = df[df['countries'].str.contains(',')]
display(df_multiple_countries)
```
