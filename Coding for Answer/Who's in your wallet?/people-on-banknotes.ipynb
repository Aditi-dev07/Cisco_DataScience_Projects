# People on Banknotes

Whose faces appear on banknotes?

The file `people-on-banknotes.csv` contains data about individuals featured on banknotes from 38 countries. This dataset spans all 22 subregions and sub-subregions of the world, as defined by the United Nations Statistics Division's geoscheme.

It profiles 241 people, detailing their occupations and the year they first appeared on a banknote. Additionally, it includes their year of death — or `NaN` if they were still alive when the dataset was compiled.

Most banknotes were issued after the featured individual’s death. The column `first_death_diff` calculates the difference between the year of their first appearance on a banknote and their year of death (or remains `NaN` if the person was still living at the time of curation).

```python
import pandas as pd

df = pd.read_csv('people-on-banknotes.csv')
df
```

### Quick cleaning

The same person can appear on multiple banknotes. Below we drop the `value` column and remove duplicate people.

```python
df = df.drop(columns=['value'])
df = df.drop_duplicates(subset="name")
df
```

##1.What proportion of individuals featured are male versus female?

```python
Gender_proportion = df['gender'].value_counts(normalize=True)
Gender_proportion
```

##2.Are writers or politicians more commonly depicted

```python
print(df['occupation'].unique())
```

```python
writers_count = df.query("occupation == 'Writer'").value_counts().sum()
politicians_count = df.query("occupation == 'Politician'").value_counts().sum()
print(f"Number of writers: {writers_count}")
print(f"Number of politicians: {politicians_count}")

if writers_count > politicians_count:
    print("Writers are more commonly depicted.")
elif politicians_count > writers_count:
    print("Politicians are more commonly depicted.")
else:
    print("Writers and politicians are depicted in equal numbers.")
```

##3.What percentage of featured individuals are musicians?



```python
Musician_count = df.query('occupation == "Musician"').value_counts().sum()
print(f"Number of musicians: {Musician_count}")
print(f"Percentage of musicians: {Musician_count / len(df) * 100}")
```

##4.What percentage of banknotes were issued before the person’s death?

```python
before_death = df.query('first_death_diff < 0').value_counts().sum()
print(f"Number of banknotes issued before the person’s death: {before_death}")
print(f"Percentage of banknotes issued before the person’s death: {before_death / len(df) * 100}")
```

##5.Who is the oldest historical figure in the dataset?

```python
oldest_historical_figure = df.query('first_death_diff > 0').sort_values(by='first_death_diff', ascending=False).iloc[0]
print(oldest_historical_figure)
```

##6.Which countries feature the oldest historical figures on their banknotes?

```python
oldest_historical = df.query('first_death_diff > 0')
country_oldest_figures = oldest_historical.groupby('country')['first_death_diff'].median().sort_values(ascending=False)
print(country_oldest_figures)
```

##7.What percentage of individuals died at least 100 years before appearing on a banknote

```python
individuals_died = df.query('first_death_diff >= 100')
print(f"Number of individuals died at least 100 years before appearing on a banknote: {len(individuals_died)}")
print(f"Percentage of individuals died at least 100 years before appearing on a banknote: {len(individuals_died) / len(df) * 100}")
```

##8.Which individuals appeared on a banknote just one year after their death?

```python
individuals_appeared = df.query('first_death_diff == 1')
print(individuals_appeared[['name', 'country', 'gender', 'occupation']])
```
